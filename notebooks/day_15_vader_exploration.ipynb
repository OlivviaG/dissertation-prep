''' Sentiment Analysis ! '''

'1. I've just had the absolute worst day, it could not have been any worse.' 
'2. Today wasn't the best, I felt a little down and lonely.' 
'3. It was a boring but relaxed day today.' 
'4. This was a really lovely afternoon, I had a nice time.' 
'5. The sun is out, the birds are singing and I feel estatic!'

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

analyser = SentimentIntensityAnalyzer()

phrases = ["1. I've just had the absolute worst day, it could not have been any worse.", 
           "2. Today wasn't the best, I felt a little down and lonely.", 
           "3. It was a boring but relaxed day today.", 
           "4. This was a really lovely afternoon, I had a nice time.", 
           "5. The sun is out, the birds are singing and I feel ecstatic!"]

for phrase in phrases:
    scores = analyser.polarity_scores(phrase)
    print(phrase + "\n", scores, "\n")
```
